```python
import synapseclient
import pandas as pd
syn = synapseclient.Synapse()
syn.login()
```

## Annotate new assets

```python
succeeded = pd.read_csv('tmp/succeeded.csv')

for i in range(len(succeeded)):
    synid = succeeded.iloc[i,0]
    key = succeeded.iloc[i,1]
    value = succeeded.iloc[i,3]
    annos = syn.get_annotations(synid)
    annos[key] = value
    syn.set_annotations(annos)
```

## Annotate all assets



```python
latest_assets = pd.read_csv('htan-assets-latest.csv')

for i in range(len(latest_assets)):
#for i in range(4):
    synid = latest_assets.iloc[i,2]
    key = latest_assets.iloc[i,4]
    value = latest_assets.iloc[i,7]
    annos = syn.get_annotations(synid)
    annos[key] = value
    syn.set_annotations(annos)
```

## Get component of manifests by synapse id



```python
Manifests = syn.tableQuery("SELECT * FROM syn20446927 WHERE (\"name\"='synapse_storage_manifest.csv')").asDataFrame()
```

```python
import os
import logging
from tqdm import tqdm
logging.disable(logging.DEBUG)

ImagingLevel2 = []

for index, row in Manifests.iterrows():
    syn.get(row.id, downloadLocation = 'local_synapse_storage_manifests')
    manifest_df = pd.read_csv('local_synapse_storage_manifests/synapse_storage_manifest.csv')
    try:
        component = manifest_df['Component'].values[0]
    except:
        os.remove('local_synapse_storage_manifests/synapse_storage_manifest.csv')
        continue
    if component == "ImagingLevel2":
        print(row.id + ": ImagingLevel2")
        ImagingLevel2.append(manifest_df)
    os.remove('local_synapse_storage_manifests/synapse_storage_manifest.csv')

ImagingLevel2 = pd.concat(ImagingLevel2)
ImagingLevel2
```

```python
ImagingLevel2

ometiffs = ImagingLevel2[ImagingLevel2.Filename.str.contains('\.ome.tif{1,2}$')].copy()
print(len(ometiffs))
ometiffs = ometiffs[~ometiffs['HTAN Participant ID'].str.contains('HTA1_')].copy()
print(len(ometiffs))
ometiffs.rename(columns = {'entityId':'synapseId'}, inplace = True)
ometiffs['minerva'] = ''
ometiffs['thumbnail'] = ''
ometiffs.to_csv('ome_tiff_synids.csv',columns=['synapseId', 'minerva', 'thumbnail'], header=True, index = False)
```

```python
synstage_test = ometiffs.sample(n=  5).copy()
synstage_test['export'] = "syn://" + synstage_test['synapseId']
synstage_test.to_csv('synstage_test.csv', columns = ['export'], header = False, index= False)
```
